# What is Machine Learning?

#### *Creating models that help us understand and interpret previously observed data, and also that allow us to make predictions using newly observed data.*

## Types of machine learning

### *Supervised*

Modelling the relationship between some measured "features" of data and a "label" associated with that data
- **Categorical** --> label contains two or more discrete categories
- **Regression** --> label is continuous

### *Unsupervised*

Modeling features **without** a label
- **Clustering** --> finding distinct groups of data
- **Dimensionality Reduction** --> finding more succinct representations

## What is a Feature?

A **Feature** is an individual measurable property or characteristic of a dataset.

### Feature Types
- Numerical
- Continuous

#### Numerical
- Continuous values measured on a scale
- Examples: age, temperature, price

#### Categorical
- Discrete values grouped into categories
- Typically **encoded into numerical features** for ML algorithms
- Some algorithms natively support both types (e.g., random forest, decision trees)
- Examples: gender, color, zip code

---

**Note:** Feature construction can influence modeling accuracy and general understanding of structure, particularly in high-dimensional problems.

## Formatting Data

For machine learning, data is commonly processed in the form of a table.

- **Rows** → elements or "samples" (`n_samples` is number of rows)
- **Columns** → information describing each sample or "features" (`n_features` is number of columns)

### Features Matrix (2D)
- Shape: `[n_samples, n_features]`
- Commonly stored in a DataFrame
- Referred to as `X`
- A **feature vector** is an n-dimensional row vector of numerical features that represent some object/sample

### Label / Target Array (1D)
- Commonly stored as a Series
- Quantity we are trying to predict (aka dependent variable)
- Referred to as `y`

## Estimator API

1. **Choose a model class** — import the appropriate estimator class from Scikit-Learn

2. **Choose hyperparameters** — instantiate the class with desired values (user-specified options)

   
3. **Arrange data** — structure into a features matrix (`X`) and target vector (`y`)

   
4. **Fit the model** — call `fit()` method on the model instance with your data

   
5. **Apply the model to new data**:
   - Supervised learning: predict labels for unknown data using `predict()`
   - Unsupervised learning: transform or infer properties using `transform()` or `predict()`

### Train/Test Split
Use `train_test_split` to divide data into training and testing sets — avoid using test data when fitting the model.
