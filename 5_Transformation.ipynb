# Transformation

Feature transformation is a part of **feature engineering** where we change the *shape, scale, or structure* of current data to make it more suitable for machine learning models.

 ##### Why is it important?

 *Imagine* comparing age (0-100) and salary ($50,000-$200,000) in a data set

 Salary has *way* bigger number, machine learning models don't understand this

 So you want to fix this using different types of transformation!

### Main types of transformation:
1. **Min-Max Scaling** 
2. **Standardization (Z-Score)**
3. **Log Transform**
4. **Square Root Transform** 
5. **Box-Cox Transform**
6. **Binning (Discretization)**
7. **Binarization**

#### Example Data

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from scipy.stats import boxcox

# Our example dataset
students = pd.DataFrame({
    'name': ['Dahlia', 'Jasmine', 'Lily', 'Rose'],
    'hours_studied': [2, 5, 8, 10],
    'income': [20000, 55000, 120000, 200000],
    'exam_score': [55, 78, 95, 98],
    'age': [18, 22, 35, 45]
})
students
```

### 1. Min-Max Scaling

**Formula:** `(x - min) / (max - min)`

**When to use:** When features are on different scales and you need them all between 0 and 1. Common for neural networks and distance-based models like KNN.

```python
scaler = MinMaxScaler()
students['exam_minmax'] = scaler.fit_transform(students[['exam_score']])
print(students[['name', 'exam_score', 'exam_minmax']])
```

### 2. Standardization (Z-Score)

**Formula:** `(x - mean) / std`

**When to use:** When your model assumes normally distributed data, like linear regression or logistic regression. Centers data around 0.

```python
std_scaler = StandardScaler()
students['exam_zscore'] = std_scaler.fit_transform(students[['exam_score']])
print(students[['name', 'exam_score', 'exam_zscore']].round(2))
```

### 3. Log Transform

**Formula:** `log(1 + x)`

**When to use:** When data is heavily right-skewed (long tail to the right), like income or house prices. Compresses large values and spreads out small ones.

```python
students['income_log'] = np.log1p(students['income'])
print(students[['name', 'income', 'income_log']].round(2))
```

### 4. Square Root Transform

**Formula:** `√x`

**When to use:** When data is moderately skewed — not extreme enough for log. A gentler version of log transform.

```python
students['income_sqrt'] = np.sqrt(students['income'])
print(students[['name', 'income', 'income_sqrt']].round(2))
```

### 5. Box-Cox Transform

**Formula:** Automatically finds the best λ (lambda) power to normalize data.

**When to use:** When you're not sure which transform to pick — Box-Cox tries many and finds the best one. Only works on positive values. When λ ≈ 0 it behaves like log, λ = 0.5 like square root, λ = 1 means no change needed.

```python
transformed, best_lambda = boxcox(students['income'])
students['income_boxcox'] = transformed
print(students[['name', 'income', 'income_boxcox']].round(2))
print(f"\nBest lambda found: {best_lambda:.2f}")
```

### 6. Binning (Discretization)

**Formula:** Assign each value to a bin based on defined ranges.

**When to use:** When the exact number matters less than the group. Common for age groups, income brackets, experience levels.

```python
students['age_group'] = pd.cut(
    students['age'],
    bins=[0, 20, 30, 50],
    labels=['Young', 'Mid', 'Senior']
)
print(students[['name', 'age', 'age_group']])
```

### 7. Binarization

**Formula:** If x ≥ threshold → 1, else → 0

**When to use:** When you only care about above or below a cutoff, like pass/fail or purchased/didn't purchase.

```python
students['passed'] = (students['exam_score'] >= 70).astype(int)
print(students[['name', 'exam_score', 'passed']])
```

#### Summary

| Type | What It Does | Good For |
|------|-------------|----------|
| **Min-Max Scaling** | Values → 0 to 1 | Neural networks, KNN |
| **Standardization** | Values → mean=0, std=1 | Linear/logistic regression |
| **Log Transform** | Compresses large values | Heavily right-skewed data |
| **Square Root** | Mildly compresses values | Moderately skewed data |
| **Box-Cox** | Finds best transform automatically | When unsure which to use |
| **Binning** | Numbers → categories | Age groups, income brackets |
| **Binarization** | Values → 0 or 1 | Pass/fail, above/below cutoff |
