## Feature Engineering with the Titanic Dataset

```python
import seaborn as sns
import pandas as pd
import numpy as np

# load the dataset
df = sns.load_dataset('titanic')
```

### PART 1: IMPUTATION (15 pts)

```python
# 1. (5 pts) print missing values for each column
print('Missing values per column:')
df.isnull().sum()
```

```python
# 2. (5 pts) impute age with median
median_age = df['age'].median()
df['age'] = df['age'].fillna(median_age)
```

```python
# 3. (5 pts) impute deck with mode
mode_deck = df['deck'].mode()[0]  # mode() returns a series, so grab first value
df['deck'] = df['deck'].fillna(mode_deck)
df.isnull().sum()
```

### PART 2: FEATURE SELECTION (10 pts)

```python
# 1. (10 pts) drop one of any feature pairs with > 75% correlation

# get numerical columns only
df_num = df.select_dtypes(include=np.number)
```

```python
# compute correlation matrix
corr_matrix = df_num.corr().abs()
```

```python
# use upper triangle to avoid duplicates and self-correlation
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# find the pair with greatest correlation
max_corr = upper_tri.max().max()
max_pair = upper_tri.stack().idxmax()
print(f'{max_pair}')
# NOTE: WELCOME STUDENTS TO SELECT THE FEATURE MANUALLY TOO
```

```python
# drop one of the pair (drop the second one)
df = df.drop(columns=[max_pair[1]])
df
```

### PART 3: FEATURE CREATION (20 pts)

```python
# 1. (5 pts) create family_size column
df['family_size'] = df['sibsp'] + df['parch'] + 1
```

```python
# 2. (5 pts) create is_alone binary column
df['is_alone'] = df['alone'].apply(lambda x: 1 if x == True else 0)
```

```python
# 3. (5 pts) bin age into age_group
df['age_group'] = pd.cut(df['age'], bins=[0, 18, 60, 100], labels=['child', 'adult', 'senior'])
```

```python
# 4. (5 pts) one-hot encode pclass, sex and age_group
df = pd.get_dummies(df, columns=['pclass', 'sex', 'age_group'], dtype=int)
```

### PART 4: NORMALIZATION (15 pts)

```python
# 1. (5 pts) select numerical columns
df_num = df.select_dtypes(include="number")
```

```python
# 2. (5 pts) apply min-max normalization manually
df_normalized = (df_num - df_num.min()) / (df_num.max() - df_num.min())
```
